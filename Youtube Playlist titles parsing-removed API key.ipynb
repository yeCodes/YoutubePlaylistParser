```python
# BOILERPLATE CODE.
#https://stackoverflow.com/questions/18804904/retrieve-all-videos-from-youtube-playlist-using-youtube-v3-api


from apiclient.discovery import build

def fetch_all_youtube_videos(playlistId):
    """
    Fetches a playlist of videos from youtube
    We splice the results together in no particular order

    Parameters:
        parm1 - (string) playlistId
    Returns:
        playListItem Dict
    """
    api_key = "AIzaSyDOSvLn5HKCNqorxzFvpZZJbdp7xSpHXPc"    # key required to access API like key for door!!

    youtube = build('youtube','v3',developerKey = api_key)

    res = youtube.playlistItems().list(
    part="snippet, contentDetails", 
    playlistId=playlistId,
    maxResults="50"
    ).execute()

    nextPageToken = res.get('nextPageToken')
    while ('nextPageToken' in res):
        nextPage = youtube.playlistItems().list(
        part="snippet",
        playlistId=playlistId,
        maxResults="50",
        pageToken=nextPageToken
        ).execute()
        res['items'] = res['items'] + nextPage['items']

        if 'nextPageToken' not in nextPage:
            res.pop('nextPageToken', None)
        else:
            nextPageToken = nextPage['nextPageToken']

    return res
```

```python
outputDurations = []

for i in videoIds:
    vidDetails = fetch_video_details(i)
    duration = vidDetails['items'][0]['contentDetails']['duration']

    #print (duration)
    time = duration.split("M")
    mins = time[0]
    secs = time[1]
    
    # this is lsow because it needs to search
    mins = mins.replace("PT",'')
    secs = secs[:-1]                     #https://stackoverflow.com/questions/15478127/remove-final-character-from-string
    
    # cast as int
    mins = int(mins)
    secs = int(secs)
    
    durationString =''
    
    if secs >= 30:
        durationString = str(mins + 1) +'mins'
    else:
        durationString = str(mins) +'mins'
        
    outputDurations.append(durationString)
        
#minutes = 
#seconds = 
# SPLIT STRING ON M in to 2

#remove S convert last
#remove PT
```
